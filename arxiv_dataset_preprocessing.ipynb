## get subgroups

where every element in subgroups is an article and an article is piled up with various subgroup

```python
subgroups = [[subgroup(title=n,text=''.join(c)) for n,c in zip(name,content)] for name,content in zip(df["section_names"].to_list(),df["sections"].to_list())]
```

```python
lengths = []
titles = []
def align_title_length(subgroups):
    for article in subgroups:
        length = []
        title = []
        for subgroup in article:
            length.append(len(subgroup.text))
            title.append(assgin_prompts(summary_prompts=summary_prompts,query_title="## " + subgroup.title,replace=False)[0])
        lengths.append(length)
        titles.append(title)
    return lengths,titles

arxiv_dataset_length, arxiv_dataset_title = align_title_length(subgroups=subgroups)
arxiv_dataset_length[0],arxiv_dataset_title[0]
```

```python
def concat_pd(subtitles, 
              lengths, 
              file_names,
              prompts):
    df = pd.DataFrame(columns=list(prompts.keys()) + ['avg_length'])
    print("length:",len(subtitles),len(lengths),len(file_names))

    for file_name,file_subtitles,file_lengths in zip(file_names,subtitles,lengths):
        for subtitle,length  in zip(file_subtitles,file_lengths):
            if file_name not in df.index:   # Add a new row for 'file_name' if it does not exist yet
                df.loc[file_name] = [np.nan]*len(df.columns)
            if pd.isnull(df.loc[file_name,subtitle]):   # Use pandas .isnull() to check for None, NaN etc.
                df.loc[file_name,subtitle] = length
            else:
                df.loc[file_name,subtitle] += length
        df.loc[file_name,'avg_length'] = sum(file_lengths)/len(file_lengths)

    return df


arxiv_dataset_df = concat_pd(
     subtitles= arxiv_dataset_title,
     lengths= arxiv_dataset_length,
     file_names= df["article_id"].to_list(),
     prompts= summary_prompts
)
```

```python
rename_map = {
    "intro":"Introduction",
    "abs":"Abstract",
    "background":"Background",
    "related work":"Related Work",
    "model":"Model",
    "experiment":"Experiment",
    "conclusion":"Conclusion",
    "method":"Method",
    "discussion":"Discussion",
    "future":"Future Work",
    "limit":"Limitation",
    "dataset":"Dataset",
}
arxiv_dataset_df = arxiv_dataset_df.rename(columns=rename_map).sort_values(by='avg_length',ascending=False)
arxiv_dataset_df
```
